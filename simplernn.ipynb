```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.datasets import imdb
```

```python
train=pd.read_csv('c-b.csv')
```

```python
train.head()
```

```python
train=train.drop(columns=['Unnamed: 2'],axis=1)
```

```python
train
```

```python
train.columns
```

```python
max_features=10000 # vocab_size
user=[]
for row in train.loc[:,'User']:
    user.append(row)
user
```

```python
voc_size=10000
one_hot_repr=[one_hot(words,voc_size) for words in user]
one_hot_repr
```

```python
sent_len=15
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_len) 
```

```python
embedded_docs
```

```python
dim=10
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_len))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
